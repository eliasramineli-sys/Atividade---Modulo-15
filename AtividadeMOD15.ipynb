```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando a base de dados
df = pd.read_csv(r"C:\Users\ELIAS\Desktop\Modulo 14\dados_tratado.csv", delimiter=';')

# Visualizando os primeiros dados
print(df.head())
```

# 2) Comece pela análise univariada:

B) Já é possível identificar variáveis com possíveis outliers? Se sim, quais?

```python
# ============================================================
# 2) ANÁLISE UNIVARIADA
# ============================================================

# A) Estatísticas descritivas das variáveis numéricas
print("A) Estatísticas descritivas das variáveis numéricas:")
display(df.describe().T)

# Estatísticas para colunas categóricas/object
print("\nEstatísticas das variáveis categóricas:")
display(df.describe(include='object').T)

# ------------------------------------------------------------
# B) Identificação inicial de possíveis outliers pelo método IQR
# ------------------------------------------------------------
print("\nB) Possíveis variáveis com outliers pelo método IQR:")

colunas_numericas = df.select_dtypes(include=['int64', 'float64']).columns
resumo_outliers = []

for coluna in colunas_numericas:
    q1 = df[coluna].quantile(0.25)
    q3 = df[coluna].quantile(0.75)
    iqr = q3 - q1

    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr

    qtd_outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)].shape[0]

    if qtd_outliers > 0:
        resumo_outliers.append({
            'Coluna': coluna,
            'Q1': q1,
            'Q3': q3,
            'IQR': iqr,
            'Limite Inferior': limite_inferior,
            'Limite Superior': limite_superior,
            'Qtd Outliers': qtd_outliers,
            '% Outliers': round((qtd_outliers / len(df)) * 100, 2)
        })

outliers_df = pd.DataFrame(resumo_outliers)
display(outliers_df)

print("""
Insight:
As variáveis listadas acima possuem valores fora dos limites calculados pelo método IQR.
Esses valores podem ser erros, casos extremos reais ou clientes com comportamento muito diferente da maioria.
Antes de remover qualquer outlier, é importante analisar se ele faz sentido para o negócio.
""")

# ------------------------------------------------------------
# C) Gráficos importantes para análise univariada
# ------------------------------------------------------------

# Variável preditora: Churn
print("Distribuição da variável Churn:")
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='Churn')
plt.title('Distribuição da variável Churn')
plt.xlabel('Churn')
plt.ylabel('Quantidade')
plt.show()

print("""
Insight:
Este gráfico mostra se a variável alvo Churn está balanceada.
Se existir muito mais clientes que não cancelaram do que clientes que cancelaram,
o modelo pode ter tendência a aprender melhor a classe majoritária.
""")

# Seleciona automaticamente algumas colunas numéricas para histogramas
variaveis_para_grafico = list(colunas_numericas[:4])

for coluna in variaveis_para_grafico:
    plt.figure(figsize=(7, 4))
    sns.histplot(df[coluna], kde=True)
    plt.title(f'Distribuição da variável {coluna}')
    plt.xlabel(coluna)
    plt.ylabel('Frequência')
    plt.show()

    print(f"""
Insight sobre {coluna}:
O histograma permite observar concentração dos dados, assimetria e possíveis valores extremos.
Quando a distribuição fica muito puxada para um lado, pode indicar assimetria.
Quando aparecem barras muito isoladas nas pontas, pode indicar possíveis outliers.
""")

# Boxplot das mesmas variáveis
for coluna in variaveis_para_grafico:
    plt.figure(figsize=(7, 3))
    sns.boxplot(data=df, x=coluna)
    plt.title(f'Boxplot da variável {coluna}')
    plt.xlabel(coluna)
    plt.show()

    print(f"""
Insight sobre {coluna}:
O boxplot ajuda a visualizar mediana, dispersão e possíveis outliers.
Pontos fora das extremidades do gráfico indicam valores extremos que precisam ser avaliados.
""")

# ------------------------------------------------------------
# D) Verificação de balanceamento das variáveis booleanas
# ------------------------------------------------------------
print("\nD) Balanceamento das variáveis booleanas:")

colunas_booleanas = []

for coluna in df.columns:
    valores_unicos = df[coluna].dropna().unique()
    qtd_unicos = len(valores_unicos)

    if qtd_unicos == 2:
        colunas_booleanas.append(coluna)

for coluna in colunas_booleanas:
    print(f"\nColuna: {coluna}")
    display(df[coluna].value_counts(normalize=True).mul(100).round(2).astype(str) + '%')

    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=coluna)
    plt.title(f'Balanceamento da variável {coluna}')
    plt.xlabel(coluna)
    plt.ylabel('Quantidade')
    plt.show()

print("""
Conclusão:
Uma variável booleana é considerada desbalanceada quando uma das categorias aparece muito mais do que a outra.
Esse desbalanceamento pode influenciar análises e modelos preditivos.
""")
```

# 3) Identifique e trate as colunas que contém outliers.
Caso opte por mante-los ou altera-los justifique sua escolha.


```python
# ============================================================
# 3) IDENTIFICAÇÃO E TRATAMENTO DE OUTLIERS
# ============================================================

# Neste exercício, vamos tratar outliers usando Winsorização pelo método IQR.
# Isso significa que valores abaixo do limite inferior serão substituídos pelo limite inferior,
# e valores acima do limite superior serão substituídos pelo limite superior.
#
# Justificativa:
# Em problemas de churn, valores extremos podem representar clientes reais.
# Por isso, em vez de excluir linhas, é mais seguro limitar os valores extremos,
# evitando perda de informação importante.

df_tratado = df.copy()

tratamento_outliers = []

for coluna in colunas_numericas:
    q1 = df_tratado[coluna].quantile(0.25)
    q3 = df_tratado[coluna].quantile(0.75)
    iqr = q3 - q1

    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr

    qtd_antes = df_tratado[(df_tratado[coluna] < limite_inferior) | 
                           (df_tratado[coluna] > limite_superior)].shape[0]

    if qtd_antes > 0:
        df_tratado[coluna] = np.where(df_tratado[coluna] < limite_inferior, limite_inferior, df_tratado[coluna])
        df_tratado[coluna] = np.where(df_tratado[coluna] > limite_superior, limite_superior, df_tratado[coluna])

    qtd_depois = df_tratado[(df_tratado[coluna] < limite_inferior) | 
                            (df_tratado[coluna] > limite_superior)].shape[0]

    tratamento_outliers.append({
        'Coluna': coluna,
        'Outliers Antes': qtd_antes,
        'Outliers Depois': qtd_depois,
        'Tratamento': 'Winsorização IQR' if qtd_antes > 0 else 'Sem tratamento necessário'
    })

tratamento_outliers_df = pd.DataFrame(tratamento_outliers)
display(tratamento_outliers_df)

print("""
Conclusão do tratamento:
Foi criada uma nova base chamada df_tratado para preservar a base original df.
Os outliers foram limitados pelos limites do IQR, sem excluir clientes da base.
Essa abordagem é indicada quando os valores extremos podem ser reais e importantes para o negócio.
""")

# Comparação visual antes e depois para as colunas que tinham outliers
colunas_com_outliers = tratamento_outliers_df[tratamento_outliers_df['Outliers Antes'] > 0]['Coluna'].tolist()

for coluna in colunas_com_outliers[:5]:
    plt.figure(figsize=(7, 3))
    sns.boxplot(x=df[coluna])
    plt.title(f'Antes do tratamento - {coluna}')
    plt.show()

    plt.figure(figsize=(7, 3))
    sns.boxplot(x=df_tratado[coluna])
    plt.title(f'Depois do tratamento - {coluna}')
    plt.show()
```

B) Quais variáveis você acredita serem as mais importantes para esse projetos relacionadas a variável Churn?

```python
# ============================================================
# 4) ANÁLISE BIVARIADA
# ============================================================

# A análise bivariada compara duas variáveis.
# Como o objetivo é entender o Churn, vamos comparar outras variáveis com Churn.

# Para evitar erro caso Churn esteja escrito com diferença de maiúsculas/minúsculas:
coluna_churn = 'Churn'

# ------------------------------------------------------------
# Pergunta 1: Clientes com maior pagamento mensal cancelam mais?
# ------------------------------------------------------------
possiveis_pagamento = [col for col in df.columns if 'mensal' in col.lower() or 'monthly' in col.lower()]
if possiveis_pagamento:
    col_pagamento = possiveis_pagamento[0]

    plt.figure(figsize=(7, 4))
    sns.boxplot(data=df_tratado, x=coluna_churn, y=col_pagamento)
    plt.title(f'{col_pagamento} x Churn')
    plt.xlabel('Churn')
    plt.ylabel(col_pagamento)
    plt.show()

    print(f"""
Pergunta 1: Clientes com maior {col_pagamento} cancelam mais?
Insight:
Se a mediana de pagamento mensal for maior no grupo de Churn = Yes/Sim/1,
isso pode indicar que clientes com mensalidades maiores têm maior tendência de cancelamento.
""")

# ------------------------------------------------------------
# Pergunta 2: Clientes com maior tempo de contrato cancelam menos?
# ------------------------------------------------------------
possiveis_tempo = [col for col in df.columns if 'tenure' in col.lower() or 'tempo' in col.lower()]
if possiveis_tempo:
    col_tempo = possiveis_tempo[0]

    plt.figure(figsize=(7, 4))
    sns.boxplot(data=df_tratado, x=coluna_churn, y=col_tempo)
    plt.title(f'{col_tempo} x Churn')
    plt.xlabel('Churn')
    plt.ylabel(col_tempo)
    plt.show()

    print(f"""
Pergunta 2: Clientes com maior {col_tempo} cancelam menos?
Insight:
Normalmente, clientes mais antigos tendem a cancelar menos.
Se o grupo que cancelou tiver menor tempo de contrato, essa variável pode ser muito importante para o modelo.
""")

# ------------------------------------------------------------
# Pergunta 3: O tipo de contrato influencia o churn?
# ------------------------------------------------------------
possiveis_contrato = [col for col in df.columns if 'contrato' in col.lower() or 'contract' in col.lower()]
if possiveis_contrato:
    col_contrato = possiveis_contrato[0]

    tabela_contrato = pd.crosstab(df[col_contrato], df[coluna_churn], normalize='index') * 100
    display(tabela_contrato.round(2))

    tabela_contrato.plot(kind='bar', figsize=(8, 4))
    plt.title(f'Tipo de contrato x Churn')
    plt.xlabel(col_contrato)
    plt.ylabel('%')
    plt.xticks(rotation=45)
    plt.show()

    print("""
Pergunta 3: O tipo de contrato influencia o churn?
Insight:
Se contratos mensais apresentarem maior percentual de churn,
isso indica que clientes com menor vínculo contratual são mais propensos a cancelar.
""")

# ------------------------------------------------------------
# Pergunta 4: Método de pagamento influencia o churn?
# ------------------------------------------------------------
possiveis_pagamento_tipo = [col for col in df.columns if 'payment' in col.lower() or 'pagamento' in col.lower()]
possiveis_pagamento_tipo = [col for col in possiveis_pagamento_tipo if col != possiveis_pagamento[0] if possiveis_pagamento]
if possiveis_pagamento_tipo:
    col_metodo_pagamento = possiveis_pagamento_tipo[0]

    tabela_pagamento = pd.crosstab(df[col_metodo_pagamento], df[coluna_churn], normalize='index') * 100
    display(tabela_pagamento.round(2))

    tabela_pagamento.plot(kind='bar', figsize=(8, 4))
    plt.title(f'Método de pagamento x Churn')
    plt.xlabel(col_metodo_pagamento)
    plt.ylabel('%')
    plt.xticks(rotation=45)
    plt.show()

    print("""
Pergunta 4: O método de pagamento influencia o churn?
Insight:
Alguns métodos de pagamento podem estar associados a clientes com maior risco de cancelamento.
Isso pode ajudar a empresa a criar estratégias específicas de retenção.
""")

# ------------------------------------------------------------
# Pergunta 5: Serviços adicionais influenciam no churn?
# ------------------------------------------------------------
colunas_categoricas = df.select_dtypes(include='object').columns.tolist()
colunas_categoricas = [col for col in colunas_categoricas if col != coluna_churn]

if colunas_categoricas:
    col_servico = colunas_categoricas[0]

    tabela_servico = pd.crosstab(df[col_servico], df[coluna_churn], normalize='index') * 100
    display(tabela_servico.round(2))

    tabela_servico.plot(kind='bar', figsize=(8, 4))
    plt.title(f'{col_servico} x Churn')
    plt.xlabel(col_servico)
    plt.ylabel('%')
    plt.xticks(rotation=45)
    plt.show()

    print(f"""
Pergunta 5: A variável {col_servico} influencia o churn?
Insight:
Ao comparar percentuais de churn entre as categorias, conseguimos identificar quais grupos de clientes
apresentam maior risco de cancelamento.
""")

# ------------------------------------------------------------
# B) Variáveis mais importantes relacionadas ao Churn
# ------------------------------------------------------------
print("""
B) Variáveis que provavelmente são mais importantes para o projeto de Churn:

1. Tempo de contrato/tenure:
Clientes mais novos geralmente têm maior chance de cancelar.

2. Tipo de contrato:
Contratos mensais normalmente apresentam maior churn do que contratos anuais ou de dois anos.

3. Pagamento mensal:
Valores mensais mais altos podem aumentar a chance de cancelamento.

4. Método de pagamento:
Alguns métodos podem estar associados a maior risco de churn.

5. Serviços contratados:
A presença ou ausência de serviços adicionais pode influenciar a permanência do cliente.

Conclusão geral:
As variáveis mais importantes são aquelas que apresentam diferenças claras entre os grupos de clientes
que cancelaram e os que permaneceram.
""")
```
